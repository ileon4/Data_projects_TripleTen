## Feature Encoding

### Encoding datetime features

Since ML algorithms can't work with datetime data types, I need to use ordinal encoding in order to create buckets signaling weights to each bucket in these features.

First let's get an idea for the range of dates in each of these 3 columns.

```python
# DateCrawled - date profile was downloaded from the database
print('oldest:', df['DateCrawled'].min(),'most recent:', df['DateCrawled'].max())
```

```python
# DateCreated — date of profile creation
print('oldest:', df['DateCreated'].min(),'most recent:', df['DateCreated'].max())
```

```python
# LastSeen — date of the last activity of the user
print('oldest:', df['LastSeen'].min(),'most recent:', df['LastSeen'].max())
```

I will create a function outlining the datetime buckets to apply to all 3 features that need encoding. For that I need a single reference date to go off of for all 3 features. I am using this strategy to ensure consistency and simplicity.

```python
# Use the overall maximum date across all datetime columns
reference_date = max(df['DateCrawled'].max(), df['DateCreated'].max(), df['LastSeen'].max())
print(f"Reference date: {reference_date}")
```

```python
# Keep your existing bucketed approach (with fixed mapping)
# PLUS add raw differences
df['days_since_crawled'] = (reference_date - df['DateCrawled']).dt.days
df['days_since_created'] = (reference_date - df['DateCreated']).dt.days
df['days_since_lastseen'] = (reference_date - df['LastSeen']).dt.days
```

```python
def categorize_listing_age(date_value):
    days_diff = (reference_date - date_value).days
    
    if days_diff <= 30:
        return 'very_recent'
    elif days_diff <= 90:
        return 'recent'  
    elif days_diff <= 365:
        return 'within_year'
    else:
        return 'old'

# Then apply ordinal encoding
crawled_dict = {'very_recent': 3, 'recent': 2, 'within_year': 1, 'old': 0}
lastseen_dict = {'very_recent': 3, 'recent': 2, 'within_year': 1, 'old': 0}  
created_dict = {'very_recent': 3, 'recent': 2, 'within_year': 1, 'old': 0}
```

```python
df['DateCrawled'] = df['DateCrawled'].apply(categorize_listing_age)
df['DateCreated'] = df['DateCreated'].apply(categorize_listing_age)
df['LastSeen'] = df['LastSeen'].apply(categorize_listing_age)
```

```python
# Apply the ordinal mappings to your categorized datetime columns
df['DateCrawled'] = df['DateCrawled'].map(crawled_dict)
df['DateCreated'] = df['DateCreated'].map(created_dict)
df['LastSeen'] = df['LastSeen'].map(lastseen_dict)
```

## Splitting Data and Declaring Target and Feature Variables

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

# Define target and features
target = df['Price']  # Your target variable
features = df.drop(['Price'], axis=1)  # Your feature variables

# First split: separate test set (20% of total data)
features_temp, X_test, target_temp, y_test = train_test_split(
    features, target, test_size=0.2, random_state=12345
)

# Second split: divide remaining data into train/validation
features_train, features_valid, target_train, target_valid = train_test_split(
    features_temp, target_temp, test_size=0.25, random_state=12345  # 0.25 of 0.8 = 0.2 of total
)
```

### Target Encoding High Cardinality Features

For the high cardinality categorical features, one hot encoding would create too many features.  

Target Encoding (Mean Encoding):  

This replaces each category with the mean target value for that category. 

Target encoding often works well for price prediction because it directly relates each brand/model to the average price.  

Target encoding is special because it directly uses the target variable to create the encoding. This is also why we use target encoding after splitting the data.

```python
# Check data types in your current splits
print("Features_train dtypes:")
print(features_train.dtypes)
print("\nAny object columns?")
print(features_train.select_dtypes(include=['object']).columns.tolist())
```

```python
# Create copies of feature sets to avoid SettingWithCopyWarning
features_train = features_train.copy()
features_valid = features_valid.copy()
X_test = X_test.copy()

# Now apply target encoding
# Target encoding for Brand
brand_means = features_train.join(target_train).groupby('Brand')['Price'].mean()

features_train['Brand_encoded'] = features_train['Brand'].map(brand_means)
features_valid['Brand_encoded'] = features_valid['Brand'].map(brand_means)
X_test['Brand_encoded'] = X_test['Brand'].map(brand_means)

# Target encoding for Model  
model_means = features_train.join(target_train).groupby('Model')['Price'].mean()

features_train['Model_encoded'] = features_train['Model'].map(model_means)
features_valid['Model_encoded'] = features_valid['Model'].map(model_means)
X_test['Model_encoded'] = X_test['Model'].map(model_means)

# Drop original categorical columns
features_train = features_train.drop(['Brand', 'Model'], axis=1)
features_valid = features_valid.drop(['Brand', 'Model'], axis=1)
X_test = X_test.drop(['Brand', 'Model'], axis=1)

# Verify all columns are numeric
print("Any object columns remaining?")
print(features_train.select_dtypes(include=['object']).columns.tolist())
```

## Model training

### Linear Regression

```python
%%time
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model_lr = LinearRegression()
model_lr.fit(features_train, target_train) # train model on training set
predictions_valid_lr = model_lr.predict(features_valid) # get model predictions on validation set

result_lr = mean_squared_error(target_valid, predictions_valid_lr) ** 0.5
print("RMSE of the linear regression model on the validation set:", result_lr)

# cross-validation
cv_scores = cross_val_score(model_lr, features_train, target_train, 
                           cv=5, scoring='neg_mean_squared_error')
cv_rmse = (-cv_scores) ** 0.5
print(f"Cross-validation RMSE: {cv_rmse.mean():.2f} (+/- {cv_rmse.std() * 2:.2f})")
```

### Decision Tree

```python
%%time
from sklearn.tree import DecisionTreeRegressor

best_model_dt = None
best_result_dt = 10000
best_depth_dt = 0
for depth in range(1, 7): # choose hyperparameter range
    model_dt = DecisionTreeRegressor(random_state=12345, max_depth=depth)# train model on training set
    model_dt.fit(features_train, target_train) # train model on training set
    predictions_valid_dt = model_dt.predict(features_valid) # get model predictions on validation set
    result_dt = mean_squared_error(target_valid, predictions_valid_dt) ** 0.5 # calculate RMSE on validation set
    if result_dt < best_result_dt:
        best_model_dt = model_dt
        best_result_dt = result_dt
        best_depth_dt = depth

print(f"RMSE of the best model on the validation set (max_depth = {best_depth_dt}): {best_result_dt}")

# cross-validation
cv_scores = cross_val_score(best_model_dt, features_train, target_train, 
                           cv=5, scoring='neg_mean_squared_error')
cv_rmse = (-cv_scores) ** 0.5
print(f"Cross-validation RMSE: {cv_rmse.mean():.2f} (+/- {cv_rmse.std() * 2:.2f})")
```

### Random Forest

```python
%%time
from sklearn.ensemble import RandomForestRegressor

best_model_rf = None
best_result_rf = 10000
best_est_rf = 0
best_depth_rf = 0
for est in range(20, 41, 10):
    for depth in range (5, 9):
        model_rf = RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=depth)# initialize model constructor with parameters random_state=12345 and n_estimators=est
        model_rf.fit(features_train, target_train) # train model on training set
        predictions_valid_rf = model_rf.predict(features_valid) # get model predictions on validation set
        result_rf = mean_squared_error(target_valid, predictions_valid_rf) ** 0.5 # calculate RMSE on validation set
        if result_rf < best_result_rf:
            best_model_rf = model_rf
            best_result_rf = result_rf
            best_est_rf = est
            best_depth_rf = depth

print("RMSE of the best model on the validation set:", best_result_rf, "n_estimators:", best_est_rf, "best_depth:", best_depth_rf)

# cross-validation
cv_scores = cross_val_score(best_model_rf, features_train, target_train, 
                           cv=5, scoring='neg_mean_squared_error')
cv_rmse = (-cv_scores) ** 0.5
print(f"Cross-validation RMSE: {cv_rmse.mean():.2f} (+/- {cv_rmse.std() * 2:.2f})")
```

```python
%%time
import lightgbm as lgb

best_model_lgb = None
best_result_lgb = 10000
best_params = {}

# Focused ranges for faster training
for n_est in [50, 100, 150]:                    # Number of trees
    for lr in [0.05, 0.1, 0.15]:               # Learning rate
        for depth in [6, 8, 10]:               # Max depth
            model_lgb = lgb.LGBMRegressor(
                n_estimators=n_est,
                learning_rate=lr,
                max_depth=depth,
                random_state=12345,
                verbose=-1  # Suppress output
            )
            
            model_lgb.fit(features_train, target_train)
            predictions_valid_lgb = model_lgb.predict(features_valid)
            result_lgb = mean_squared_error(target_valid, predictions_valid_lgb) ** 0.5
            
            if result_lgb < best_result_lgb:
                best_model_lgb = model_lgb
                best_result_lgb = result_lgb
                best_params = {
                    'n_estimators': n_est,
                    'learning_rate': lr,
                    'max_depth': depth
                }

print(f"RMSE of the best LightGBM model: {best_result_lgb}")
print(f"Best parameters: {best_params}")

# cross-validation
cv_scores = cross_val_score(best_model_lgb, features_train, target_train, 
                           cv=5, scoring='neg_mean_squared_error')
cv_rmse = (-cv_scores) ** 0.5
print(f"Cross-validation RMSE: {cv_rmse.mean():.2f} (+/- {cv_rmse.std() * 2:.2f})")
```
